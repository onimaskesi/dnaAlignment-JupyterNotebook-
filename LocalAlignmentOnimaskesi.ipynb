```python
bosluk = '_'

sembol_eslesme = '*'
sembol_eslesmeme = '!'
sembol_bosluk = ' '

# puanlama değerleri
gap = -1
match = 10
mismatch = -2

q = list("1101001100")
p = list("110101000")

# 0 değerlerine sahip, üzerinde işlem yapacağımız bir matris oluştur.
stun_sayisi, satir_sayisi = len(q) + 1 , len(p) + 1;
Matrix_Local = [[0 for x in range(stun_sayisi)] for y in range(satir_sayisi)] 

# Buradaki fonksiyonlar üzerinde işlem yapılacak olan matris elemanının, kosegen değerini, sol değerini ve üst değerini verir
#alanlar(satır,stün): 
#köşegeni(satır-1,stün-1),
def kosegen(stirIndex,stunIndex,Matrix):
    return Matrix[stirIndex-1][stunIndex-1]

#sol(satır,stun-1)
def sol(stirIndex,stunIndex,Matrix):
    return Matrix[stirIndex][stunIndex-1] 
#üst(satır-1,stun-1)
def ust(stirIndex,stunIndex,Matrix):
    return Matrix[stirIndex-1][stunIndex]


def matrixPuanlaLocal():
    for i in range(1, satir_sayisi): #satır sayısı
        for j in range(1, stun_sayisi): #stun sayısı

            #q stunu temsil eder
            #p satırı temsil eder
            if(q[j-1] == p[i-1]):
                Matrix_Local[i][j] = kosegen(i,j,Matrix_Local) + match
            else:

                Matrix_Local[i][j] = max( sol(i,j,Matrix_Local) + gap ,\
                                         ust(i,j,Matrix_Local) + gap , \
                                         kosegen(i,j,Matrix_Local) + mismatch )  

                if(Matrix_Local[i][j] < 0):
                    Matrix_Local[i][j] = 0


matrixPuanlaLocal()
# local hizalamada, hizalama işlemine en büyük değer ile başlandığı için:
# max değeri ve indexlerini hesaplama burada yapılmıştır
enBuyukDeger = -1
satir_index_max = 0
stun_index_max = 0
for i in range(satir_sayisi): #satır sayısı
        for j in range(stun_sayisi): #stun sayısı
            if(Matrix_Local[i][j] > enBuyukDeger):
                enBuyukDeger = Matrix_Local[i][j]
                satir_index_max = i
                stun_index_max = j
                
Q_local = []
sembol_local=[]
P_local = []
# eşleşme varsa: 
# kosegene git => bulunulan konumdaki eşleşmeyi al Q.insert(0,value)
# eşleşme yoksa en yüksek değere sahip olan komşuya git komşular arasından:
# kosegene gidersen => bulunulan konumu al Q.insert(0,q[stun]) P.insert(0,p[satır])
# sola gidersen => stun değerini aynen al Q.insert(0,q[stun]), satır değerine boşluk koy P.insert(0,bosluk)
# yukarı gidersen => satır değerini aynen al P.insert(0,p[satır]), stun değerine boşluk koy Q.insert(0,bosluk)
# satır veya stun indexinin değeri 0 olunca sonlandır.
satir_indexeleri_local = []
stun_indexeleri_local = []
def hizalaLocal(satir_indexi,stun_indexi):
    
    satir_indexeleri_local.append(satir_indexi)
    stun_indexeleri_local.append(stun_indexi)
    
    if(satir_indexi != 0 and stun_indexi != 0):
        
        if(p[satir_indexi-1] == q[stun_indexi-1]):
            P_local.insert(0,p[satir_indexi-1]) # değeri ekle
            Q_local.insert(0,q[stun_indexi-1])
            sembol_local.insert(0,sembol_eslesme)
            satir_indexi -= 1 #üst
            stun_indexi -= 1 #sol
        else:
            #kosegene gidebilir( number = 1), sola gidebilir(2), ust gidebilir(3)
            enBuyuk = max(kosegen(satir_indexi,stun_indexi,Matrix_Local), \
                          sol(satir_indexi,stun_indexi,Matrix_Local), \
                          ust(satir_indexi,stun_indexi,Matrix_Local))
            
            if(enBuyuk == kosegen(satir_indexi,stun_indexi,Matrix_Local)):
                P_local.insert(0,p[satir_indexi-1])
                Q_local.insert(0,q[stun_indexi-1])
                sembol_local.insert(0,sembol_eslesmeme)
                satir_indexi -= 1 #üst
                stun_indexi -= 1 #sol
            elif(enBuyuk == sol(satir_indexi,stun_indexi,Matrix_Local)): #soldan
                Q_local.insert(0,q[stun_indexi-1])
                P_local.insert(0,bosluk)
                sembol_local.insert(0,sembol_bosluk)
                stun_indexi -= 1 #sol
            elif(enBuyuk == ust(satir_indexi,stun_indexi,Matrix_Local)): # ustten
                P_local.insert(0,p[satir_indexi-1])
                Q_local.insert(0,bosluk)
                sembol_local.insert(0,sembol_bosluk)
                satir_indexi -= 1 #üst
                
        hizalaLocal(satir_indexi,stun_indexi)


hizalaLocal(satir_index_max,stun_index_max)
print("Q: " +' '.join(Q_local))
print("   " +' '.join(sembol_local))
print("P: " +' '.join(P_local))
print("\t\t Score: " + str(enBuyukDeger))

import numpy
x_local = numpy.array(Matrix_Local)

import pandas
import pandas as pd
q.insert(0,' ')
p.insert(0,' ')
df_local = pandas.DataFrame(x_local, columns=q, index=p)


def styling_specific_cell(x,satir_indexeler,stun_indexeler):
    color = 'background-color: yellow;'
    df_styler = pd.DataFrame('', index=x.index, columns=x.columns)
    
    for i in range(len(satir_indexeler)):
        df_styler.iloc[satir_indexeler[i], stun_indexeler[i]] = color
    
    return df_styler


p01 = [" "]
for i in range(1,len(p)):
    p01.append(str(i) + " " + p[i])

q01 = [" "]
for i in range(1,len(q)):
    q01.append(str(i) + " " + q[i])
    
df01_local = pandas.DataFrame(x_local, columns=q01, index=p01)
```
